```python
import requests
from bs4 import BeautifulSoup
listURL = ["https://iuh.edu.vn/vi/don-vi-dao-tao-fi25?page=1","https://iuh.edu.vn/vi/don-vi-dao-tao-fi25?page=2"]
listGET = []
def getUrl(url):
    response = requests.get(url,verify=False)
    soup = BeautifulSoup(response.text, 'html.parser')
    target = soup.find_all("div",class_="content-info col-sm-9")
    for i in target:
        listGET.append("https://iuh.edu.vn/"+i.find("a",class_="c-i-title")['href'])
    
for i in listURL:
    getUrl(i)
```

```python
rp = requests.get("https://iuh.edu.vn/#",verify=False)
soup = BeautifulSoup(rp.text, 'html.parser')
target = soup.find_all("ul",class_="dropdown-menu")
for i in target:
    for j in i.find_all("a"):
        listGET.append("https://iuh.edu.vn/"+j['href'])
listGET[:-3]
```

```python
def getData(url):
    try:
        
        response = requests.get(url,verify=False)
        soup = BeautifulSoup(response.text, 'html.parser')
        target = soup.find('div',class_="content").text
        title = soup.find('h1').text.strip()
        return title,target
    except:
        return "",""
# lưu dữ liệu dạng list json tiếng việt
data = []
for i in listGET:
    title,content = getData(i)
    data.append({
        "title":title,
        "content":content
    })

import json
with open('data.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=4)


```
